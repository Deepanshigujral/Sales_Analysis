# calculating which month has highest earning

```python
dfresult['month']=dfresult['Order Date'].str[0:2]
```

```python
dfresult.head()
```

```python
dfresult["month"].unique()
```

```python
df1=dfresult[dfresult['month']!='Or']
```

```python
df1.head()
```

```python
df1['month']=df1['month'].astype('int32')
```

```python
df1.head()
```

```python
df1.info()
```

```python
# Convert 'Quantity Ordered' to float64
df1['Quantity Ordered'] = df1['Quantity Ordered'].astype('int32')
# Convert 'Price Each' to float64
df1['Price Each'] = df1['Price Each'].astype('float64')
```

```python
df1[['Price Each']]
```

```python
df1['Sales']=df1['Quantity Ordered']*df1['Price Each']
```

```python
df1.info()
```

```python
df1
```

```python
df1.groupby(by='month').sum().sort_values(by='Sales',ascending=False)
```

```python
# Decemeber has the highest sales
```

# Which city has the highest number of sales?
